# 数据类型

- 整数除法与浮点数除法中，“/”计算的结果都是浮点数，与除数与被除数的类型无关
- 整数除法中“//”计算的结果才是整数，浮点数除法中“//”计算的结果也是浮点数
- 浮点数的乘法存在四舍五入的精度损失，加法没有
- 数据类型转换int(), float(),没有double()
- 变量类型不固定，可以通过type()查看

```python
print(153/51) 
print(153/51.0) 
print(155/51) 
print(155/51.0) 

print(153//51) 
print(153//51.0) 
print(155//51) 
print(155//51.0) 

print(3.3*102)
print(3.3*102+1)

print(int(56.2))
print(float(56))
# print(double(56.2)) # 错误

xiaohong = "happy"
type(xiaohong)
xiaohong = 1
type(xiaohong)
```

# 编码

- Python支持中文，但计算机只能处理数字，其实只认识0和1，即二进制。如果要处理文本，就必须先把文本转换为数字才能处理。
- Unicode把所有语言都统一到一套编码里最常用的是用两个字节表示一个字符（如果要用到非常偏僻的字符，就需要4个字节）。现代操作系统和大多数编程语言都直接支持Unicode。
- 如果统一成Unicode编码，乱码问题从此消失了。但是写的文本基本上全部是英文时，用Unicode编码比ASCII编码多一倍存储空间，在存储和传输上十分不划算。
- 出现把Unicode编码转化为“可变长编码”的UTF-8编码。UTF-8编码把一个Unicode字符根据不同的数字大小编码成1～6个字节，常用的英文字母被编码成1个字节，汉字通常是3个字节，只有很生僻的字符才会被编码成4～6个字节。如果你要传输的文本包含大量英文字符，用UTF-8编码就能节省空间，ASCII编码实际上可以看成是UTF-8编码的一部分，所以只支持ASCII编码的大量历史遗留软件可以在UTF-8编码下继续工作。
  
- 使用getsizeof可以获取占据内存的字节数， 使用len可以获取变量中元素的数量

```python
import sys
a = 'hello world'
b = '你好，世界'
print(a)
print(b)
print(sys.getsizeof(a))
print(sys.getsizeof(b))
print(len(a))
print(len(b))
```
